```python
import pandas as pd
import gmaps
import citipy
import requests
import matplotlib.pyplot as plt
import numpy as np
from datetime import datetime
```

```python
#1 Generate a new set of 1500 random latitudes and longitudes 
lan = np.random.uniform(-90, 90, size = 1500)
lon = np.random.uniform(-180, 180, size = 1500)
coords = zip(lan, lon)
coords
```

```python
coordinates = list(coords)
```

```python
#Get the nearest cities with citipy 
from citipy import citipy
```

```python
#create empty list to hold cities 
cities = []
for coordinate in coordinates:
    city = citipy.nearest_city(coordinate[0], coordinate[1]).city_name
    cities.append(city)
    cities = list(set(cities))

len(cities)
```

```python
#Retrieve information from OpenWeatherMap 
from config import weather_api_key
```

```python
base_url = "http://api.openweathermap.org/data/2.5/weather?units=Imperial&APPID=" + weather_api_key
```

```python
#create empty list to hold information 
city_data = []
city_snow_data = []
record_count = 0
#retrieve respective information 
for city in cities: 
    city_url = base_url + "&q=" + city
    record_count +=1
    print(f"Processing record no. {record_count}")
    try:
        try:
            city_weather = requests.get(city_url).json()
            city_lat = city_weather["coord"]["lat"]
            city_lng = city_weather["coord"]["lon"]
            city_country = city_weather["sys"]["country"]
            city_max_temp = city_weather["main"]["temp_max"]
            city_humidity = city_weather["main"]["humidity"]
            city_clouds = city_weather["clouds"]["all"]
            city_description = city_weather["weather"][0]["description"]
            city_date = datetime.utcfromtimestamp(city_weather["dt"]).strftime('%Y-%m-%d %H:%M:%S')
            city_wind = city_weather["wind"]["speed"]
            city_rain = city_weather["rain"]["3h"]
            city_data.append({"City": city.title(),
                          "Country" :city_country,
                          "Date": city_date,
                          "Lat": city_lat,
                          "Lng": city_lng,
                          "Max Temp": city_max_temp,
                          "Humidity": city_humidity,
                          "Cloudiness": city_clouds,
                          "Wind Speed": city_wind, 
                          "Current Description": city_description,
                          "Rain inches(Last 3 hours)":city_rain})
        except (KeyError, NameError) as e:
            print(e)
            city_data.append({"City": city.title(),
                          "Country" :city_country,
                          "Date": city_date,
                          "Lat": city_lat,
                          "Lng": city_lng,
                          "Max Temp": city_max_temp,
                          "Humidity": city_humidity,
                          "Cloudiness": city_clouds,
                          "Wind Speed": city_wind, 
                          "Current Description": city_description,
                           "Rain inches(Last 3 hours)":0})
        pass
            
        city_snow = city_weather["snow"]["3h"]
        city_snow_data.append({"City": city.title(),"Snow inches (Last 3 hours)": city_snow})

    except (KeyError, NameError) as e:
        print(e)
        city_snow_data.append({"City": city.title(),"Snow inches (Last 3 hours)": 0})
print("End of Record")
```

```python
len(city_data)
len(city_snow_data)
```

```python
city_data_df = pd.DataFrame(city_data)
city_data_df.head(10)
```

```python
city_snow_datadf = pd.DataFrame(city_snow_data)
city_snow_datadf.head(10)
```

```python
#Merge the two dataframes 
complete_data_df = pd.merge(left = city_data_df, right = city_snow_datadf, how = "left", left_on= "City", right_on ="City")
complete_data_df.head(10)
```

```python
#output dataframe into csv file
output_data_file = "weather_challenge/weatherPY_challenge.csv"
complete_data_df.to_csv(output_data_file)
```
